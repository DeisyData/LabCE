```python
import matplotlib.pyplot as plt

# Datos de la lnea de tiempo
hitos = [
    {"ao": 1980, "evento": "Concepto de comunidades energticas", "tipo": "Global"},
    {"ao": 2000, "evento": "Expansin de proyectos en Europa", "tipo": "Global"},
    {"ao": 2016, "evento": "Reconocimiento en el mercado energtico", "tipo": "Global"},
    {"ao": 2014, "evento": "Ley 1715: fomento a energas no convencionales", "tipo": "Colombia"},
    {"ao": 2018, "evento": "UE impulsa comunidades energticas", "tipo": "Global"},
    {"ao": 2018, "evento": "CREG 030: autogeneracin y venta de excedentes", "tipo": "Colombia"},
    {"ao": 2020, "evento": "Proyectos piloto urbanos y rurales", "tipo": "Colombia"},
    {"ao": 2025, "evento": "Consolidacin de comunidades energticas", "tipo": "Colombia"},
]

# Colores
colores = {
    "Global": "#4BBEEC",
    "Colombia": "#452683"
}

# Extraer datos
aos = [h["ao"] for h in hitos]
eventos = [h["evento"] for h in hitos]
tipos = [h["tipo"] for h in hitos]
colores_eventos = [colores[t] for t in tipos]

# Crear figura vertical mejorada
plt.figure(figsize=(6, 8))
plt.vlines(1, min(aos)-2, max(aos)+2, color="#0F2B41", linewidth=2)  # Lnea base
plt.scatter([1]*len(aos), aos, c=colores_eventos, s=250, zorder=3, edgecolor="white", linewidth=1.5)

# Anotaciones ms cercanas a la lnea de tiempo
for i, (ao, evento, tipo) in enumerate(zip(aos, eventos, tipos)):
    lado = 0.1 if i % 2 == 0 else -0.1  # ms cerca de la lnea
    ha = 'left' if i % 2 == 0 else 'right'
    plt.text(1+lado, ao, f"{ao} - {evento}",
             ha=ha, va='center',
             fontsize=9, color="white",
             bbox=dict(boxstyle="round,pad=0.3", facecolor=colores[tipo], edgecolor="#0F2B41", alpha=0.9))

# Ajustes
plt.xticks([])
plt.yticks(aos, fontsize=9)
plt.title("Lnea de Tiempo Vertical: Hitos de las Comunidades Energticas", fontsize=12, color="#0F2B41")
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```
